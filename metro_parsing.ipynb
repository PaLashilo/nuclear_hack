```python
import json
import requests
```

```python
api_key = "456a4355-60ae-4c8e-a499-eefd523466f4"
url = f"https://apidata.mos.ru/v1/datasets/1488/rows?&filter=ObjectStatus+%3D+%27действует%27&api_key={api_key}"
r = requests.get(url)
stations = r.json()
```

```python
stations
```

```python
final_station = ["Пятницое шоссе", "Планерная", "Ховрино", "Физтех", 
                 "Алтуфьево", "Медведково", "Бульвар Рокоссовского", "Щёлковская", "Новокосино", "Котельники", 
                 "Некрасовка", "Зябликово", "Алма-Атинская", "Бульвар Дмитрия Донского", "Бунинская аллея", "Битцевский парк", 
                 "Коммунарка", "Аэропорт Внуково", "Новоясеневская", "Радиальная", "Москва-Сити","Деловой центр"]
```

```python
station_array = []

for station in stations:

    station_info = {}
    station_info["station_id"] = station["global_id"]
    station_info["station"] = station['Cells']["Station"]

    if station['Cells']["MCDStation"]:
        station_info["mcd"] = 1
    else:
        station_info["mcd"] = 0

    if station['Cells']["RailwayStation"]:
        station_info["railway_station"] = 1
    else:
        station_info["aeroexpress"] = 0

    if station['Cells']["AeroexpressStation"]:
        station_info["aeroexpress"] = 1
    else:
        station_info["aeroexpress"] = 0

    if station['Cells']["RailwayTerminal"]:
        station_info["railway_terminal"] = 1
    else:
        station_info["railway_terminal"] = 0

    if  station_info["station"] in final_station:
        station_info["end_station"] = 1
    else:
        station_info["end_station"] = 0



    station_info["line"] = station['Cells']["Line"]
    station_info["district"] = station['Cells']["District"]
    station_info["adm_area"] = station['Cells']["AdmArea"]
    station_array.append(station_info)

    
```

```python
station_array
```

```python
with open("moscow_metro.json", 'w') as fp:
    json.dump(station_array, fp=fp, ensure_ascii=False)
```
